```
from datetime import datetime, date
import math 
import matplotlib.pyplot as plt

def dol(d,m,y):
  today = date.today()
  birth = date(y,m,d)
  dol = today - birth
  return dol.days

def bioval(dol,d=0):
  dol_ = dol + d
  phyd = 23
  emod = 28
  intd = 33
  p = round(math.sin((math.pi*2)/phyd*dol_)*100,2)
  e = round(math.sin((math.pi*2)/emod*dol_)*100,2)
  i = round(math.sin((math.pi*2)/intd*dol_)*100,2)
  return p,e,i

def biorythm(d,m,y,u_name="noname"):
  u_days = dol(d,m,y)
  
  plist = []
  elist = [] 
  ilist = []
  drange = range(-5,25,1)
  for n in drange:
    plist.append(bioval(u_days,n)[0])
    elist.append(bioval(u_days,n)[1])
    ilist.append(bioval(u_days,n)[2])
  x_dates = []
  for d in drange:
    x_dates.append((date.today() + timedelta(d)).strftime("%b %d"))
  
  plt.figure(figsize=(14,2), dpi=100)
  #plt.suptitle('Biorythm of '+date.today().strftime("%B")+' for '+u_name, fontsize = 13)
  plt.suptitle('Biorythm for '+u_name, fontsize = 13)

  plt.plot(x_dates, plist, label='Phisical', color='C2')
  plt.plot(x_dates, elist, label='Emotional', color='C1')
  plt.plot(x_dates, ilist, label='Intelectual', color='C0')
  
  plt.xticks(rotation=45)
  plt.rc('xtick',labelsize=8)
  plt.rc('ytick',labelsize=8)
  plt.ylabel('BR level (%)')
  #plt.xlabel(date.today().strftime("%B"))  

  plt.grid(linestyle='--', linewidth=1, alpha=0.5)
  plt.axvline(date.today().strftime("%b %d"), color='C4') #line that indicates today
    
   
  plt.legend()
  plt.show()
```
